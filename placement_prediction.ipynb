```python
import pandas as pd

data = pd.read_csv("placement.csv")
data.head()
```

```python
data = data.iloc[:, 1:]
data.head()
```

```python
data.columns
```

```python
data
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['cgpa'],data['iq'],c=data['placement'])
```

```python
print(data)

data = data.iloc[:, 1:]
data.head()
```

```python
data.head()
```

```python
X = data.iloc[:,0:2]
Y = data.iloc[:,-1]

print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.1)
```

```python
print(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python

X_test = scaler.transform(X_test)
```

```python
X_test
```

```python

from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
 #model training
clf.fit(X_train,Y_train)
```

```python
#finding the model accuracy using the test data

# y_pred stores the output column values generated after testing the model with
# X_test
Y_pred = clf.predict(X_test)
Y_pred
```

```python
print(Y_test)
```

```python
#now lets find the accuracy of the model

from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred)
```

## So the model accuracy is 90%

## Now we wanna draw the line boundary drawn by the logictic regression algo

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_train, Y_train.values, clf=clf, legend=2)
```
